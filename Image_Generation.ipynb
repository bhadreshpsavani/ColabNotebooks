```python
from google.colab import userdata
```

```python
from huggingface_hub import InferenceClient

client = InferenceClient(
    provider="together",
    api_key=userdata.get('HF_TOKEN'),
)
```

```python
# output is a PIL.Image object
image = client.text_to_image(
    """My Plan of Sharing & Learning Cutting-Edge AI/ML Stuff Weekly
I have been working on cutting-edge tech stacks like MLOPs, GenAI, RAG multi-agent applications, and partially LLMOps, as well as on cloud platforms like GCP and AWS.
I have decided to start a quest to write about what I have learned during the week and share at least one blog or video.""",
    model="black-forest-labs/FLUX.1-dev",
)
```
